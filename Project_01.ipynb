```python
stats = pd.read_csv('player_stat.csv')
stats
```

```python
stats['Goals & Assists'] = stats['Goals'] + stats['Assists']
```

First factor we assumed is that Mancity has lots of Superstars

```python
stats.sort_values(by=['Goals & Assists'], ascending=False, inplace = True)
topstats = stats.head(100)
topstats.set_index('Name',inplace=True)
```

```python
topstats['Goals & Assists'].head(10).plot(kind='bar', figsize = (20,10))
```

We can see that only 1 player of 20/21 EPL winner is contained in this ranking. So, if it is not superstars that contributes success, what is it?

```python
import plotly.express as px
fig = px.scatter(topstats.reset_index().head(532),x = 'Club',y = 'Goals & Assists',color='Name')
fig.show()
```

```python
stats.set_index('Name',inplace=True)
```

```python
mci_passes = pd.DataFrame(stats[stats['Club']=='Manchester City'],columns=['Passes_Attempted','Perc_Passes_Completed'])
mci_passes['Passes_Complete'] = round(mci_passes['Passes_Attempted'] * mci_passes['Perc_Passes_Completed'] / 100)
mci_total_passes ={'Total Pass Complete':mci_passes['Passes_Complete'].sum(),'Mean of each Player':mci_passes['Passes_Complete'].mean()}
mci_total_passes
```

Here are some passing stats of Manchester City this season. Let's compare with middle-table team.

```python
avf_passes = pd.DataFrame(stats[stats['Club']=='Aston Villa'],columns=['Passes_Attempted','Perc_Passes_Completed'])
avf_passes['Passes_Complete'] = round(avf_passes['Passes_Attempted'] * avf_passes['Perc_Passes_Completed'] / 100)
avf_total_passes ={'Total Pass Complete':avf_passes['Passes_Complete'].sum(),'Mean of each Player':avf_passes['Passes_Complete'].mean()}
avf_total_passes
```

We can see that the Champion team passed a lot more than the middle-table team. 

```python
import plotly.express as px
fig = px.bar(pd.DataFrame(stats.groupby('Club')['Passes_Attempted'].sum().sort_values(ascending=False)).reset_index(), x='Club', y='Passes_Attempted', color='Club')
fig.show()
```

```python
Shot = pd.DataFrame(results.groupby('HomeTeam')['HS'].sum())
Shot['AS'] = results.groupby('AwayTeam')['AS'].sum()
Shot['Total'] = Shot['AS'] + Shot['HS']
Shot.sort_values(by=['Total'],ascending=False,inplace=True)
Shot = Shot.reindex(columns=['Total','HS','AS'])
```

```python
fig = px.bar(Shot.reset_index().rename(columns={"HomeTeam":"Club"}), x='Club', y=['HS','AS'])
fig.show()
```

```python
ShotOT = pd.DataFrame(results.groupby('HomeTeam')['HST'].sum())
ShotOT['AST'] = results.groupby('AwayTeam')['AST'].sum()
ShotOT['Total'] = ShotOT['AST'] + ShotOT['HST']
ShotOT.sort_values(by=['Total'],ascending=False,inplace=True)
ShotOT = ShotOT.reindex(columns=['Total','HST','AST'])
```

```python
fig = px.bar(ShotOT.reset_index().rename(columns={"HomeTeam":"Club"}), x='Club', y=['HST','AST'])
fig.show()
```

```python
ShotTTT = pd.DataFrame(Shot['Total'])
ShotTTT['Ontarget T'] = ShotOT['Total']
```

```python
ShotTTT.sort_values(by='Ontarget T', ascending=False).plot(kind='bar',figsize=(20,10))
```

### Is the age of the player one of the main keys to win the premier leaugue?

```python
df = pd.read_csv('player_stat.csv')
```

```python
standing = pd.read_csv('EPL standing.csv')
```

```python
standing['Team'] = standing['Team'].replace(['Liverpool'],'Liverpool FC')
standing['Team'] = standing['Team'].replace(['Brighton & Hove Albion'],'Brighton')
```

```python
df['Team']=df['Club']
df_merged = df.merge(standing)
df_merged['Pts'] = df_merged['Pts'].apply(pd.to_numeric)
df_merged = df_merged[df_merged['Mins']>=900]
result2 = df_merged[['Team','Age','Pos']].groupby('Team').mean('Age').sort_values(ascending = True, by='Pos')
result2['Age'].plot(kind='bar',figsize=(16,12),ylim=(23,29))
```

```python
stats['Cards'] = stats['Yellow_Cards'] + stats['Red_Cards']
a = pd.DataFrame(stats.groupby('Club')['Cards'].sum()).reset_index()
```

```python
a.sort_values(by='Cards',ascending=True,inplace=True)


default_color = "blue"
colors = {"Manchester City": "red"}
color_discrete_map = {
    c: colors.get(c, default_color) 
    for c in a['Club']}
fig = px.bar(a, x='Club', y='Cards',color='Club',color_discrete_map=color_discrete_map)
fig.update_traces(showlegend=False)


fig.show()

```
